```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
URL = 'https://www.formula1.com/en/drivers.html'
```

```python
headers = {'User-Agent':"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36"}
```

```python
drivers = open("dados_primeira_pag.html")
site = requests.get(URL, headers=headers)
soup = BeautifulSoup(drivers,'html.parser')
```

```python
### Nome do Piloto , Link da Página dele , Team dos Pilotos  e Rank 2019 ###
i=0
pilotos_2021 = []
for piloto in soup.find_all('a',attrs={'class':'listing-item--link'}):
    dados_pilotos = []
    linkp = piloto['href']
    nome = piloto['data-tracking']
    nome = re.search(r'.*?path.*"(.*)",',piloto['data-tracking'])
    #piloto = re.search(r'.*drivers(.*)',link['href'])
    teamp = piloto.find_all('p',attrs={'class':'listing-item--team f1--xxs f1-color--gray5'})
    teamp = re.search(r'>(.*)<',str(teamp))
    #rank = re.search(r'>(.*)<',str(piloto.find_all('div',attrs={'class':'rank'})))
    #print(rank.group(1))
    dados_pilotos.append(nome.group(1)) # Nome
    dados_pilotos.append(teamp.group(1)) # Equipe
    dados_pilotos.append('https://www.formula1.com'+linkp)# link da página
    pilotos_2021.append(dados_pilotos)
pilotos_2021_df = pd.DataFrame(pilotos_2021, columns=['nome','team','link'])
pilotos_2021_df = pilotos_2021_df.drop([19])
pilotos_2021_df = pilotos_2021_df.reset_index(drop=True)
#pilotos_2021_df.to_csv('pilotos_2021.csv')
pilotos_2021_df

```

```python
data_pilotos_df=[]
for link in pilotos_2021_df.link:
    dados_pilotos=[]
    link = requests.get(link)
    pilotos = BeautifulSoup(link.content,'html.parser')
    #print(pilotos)
    tb = pilotos.find_all('td',attrs={'class':'stat-value'})
    #print(tb)
    dados = re.search(r'<td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>, <td class="stat-value">(.*)</td>',str(tb))
#    dados_pilotos.append(dados.group(1)) # team
    dados_pilotos.append(dados.group(2)) # Nacionalidade
    dados_pilotos.append(dados.group(3)) # Podium
    dados_pilotos.append(dados.group(4)) # Total de Pontos
    dados_pilotos.append(dados.group(5)) # quantidade de corrida 
    dados_pilotos.append(dados.group(6)) # Quantos mundiais
    dados_pilotos.append(dados.group(7)) # corrida finalizada
    dados_pilotos.append(dados.group(8)) # posição mais alta
    dados_pilotos.append(dados.group(9)) # data de nascimento
    dados_pilotos.append(dados.group(10))# cidade natal
    data_pilotos_df.append(dados_pilotos)

    
data_pilotos_df = pd.DataFrame(data_pilotos_df, columns=['nacionalidade','Podium','total de pontos','quantidade de corrida','quantos mundiais','corridas finalizadas','posição mais alta','data de nascimento','cidade natal'])  

data_pilotos_df     
```

```python
pilotos_2021_df.join(data_pilotos_df)
```
